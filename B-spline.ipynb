```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# filling with the data

x = np.empty(0)
y = np.empty(0)

fl = open("test_func.txt", "r")
#fl = open("spline/test_func.txt", "r") 
for line in fl:
    tmp = line.split(" ")
    x = np.concatenate((x, [float(tmp[0])]), axis=0)
    y = np.concatenate((y, [float(tmp[1])]), axis=0)

fl.close()

x_a = np.empty(0)
y_a = np.empty(0)

fl = open("test_approx.txt", "r") 
#fl = open("spline/test_approx.txt", "r") 
for line in fl:
    tmp = line.split(" ")
    x_a = np.concatenate((x_a, [float(tmp[0])]), axis=0)
    y_a = np.concatenate((y_a, [float(tmp[1])]), axis=0)

fl.close()
```

```python
fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(111)

linestyles = ['-', '--', '-.']
grlinecolors = ['#000000', '#555555', '#aaaaaa']
linecolors = ['#000000', '#3163e1', '#ff9b2f']

ax.plot(x, y, lw=2.0, color='k', label='func')
ax.plot(x_a, y_a, lw=2.0, color='r', label='approx')


#ax.axis('equal')
plt.grid(True, linewidth = 0.5)
plt.tight_layout()
plt.legend()
plt.show()
fig.savefig("example.pdf")
```
