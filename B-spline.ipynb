```python
%matplotlib inline
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

```python
# filling with the data

x = np.empty(0)
y = np.empty(0)

fl = open("spline/test_func.txt", "r") 
for line in fl:
    tmp = line.split(" ")
    x = np.concatenate((x, [float(tmp[0])]), axis=0)
    y = np.concatenate((y, [float(tmp[1])]), axis=0)

fl.close()

x_3 = np.empty(0)
y_3 = np.empty(0)

fl = open("spline/test_approx(3).txt", "r") 
for line in fl:
    tmp = line.split(" ")
    x_3 = np.concatenate((x_3, [float(tmp[0])]), axis=0)
    y_3 = np.concatenate((y_3, [float(tmp[1])]), axis=0)

fl.close()

x_4 = np.empty(0)
y_4 = np.empty(0)

fl = open("spline/test_approx(4).txt", "r") 
for line in fl:
    tmp = line.split(" ")
    x_4 = np.concatenate((x_4, [float(tmp[0])]), axis=0)
    y_4 = np.concatenate((y_4, [float(tmp[1])]), axis=0)

fl.close()

x_5 = np.empty(0)
y_5 = np.empty(0)

fl = open("spline/test_approx(5).txt", "r") 
for line in fl:
    tmp = line.split(" ")
    x_5 = np.concatenate((x_5, [float(tmp[0])]), axis=0)
    y_5 = np.concatenate((y_5, [float(tmp[1])]), axis=0)

fl.close()

x_6 = np.empty(0)
y_6 = np.empty(0)

fl = open("spline/test_approx(6).txt", "r") 
for line in fl:
    tmp = line.split(" ")
    x_6 = np.concatenate((x_6, [float(tmp[0])]), axis=0)
    y_6 = np.concatenate((y_6, [float(tmp[1])]), axis=0)

fl.close()
```

```python
fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(111)

linestyles = ['-', '--', '-.']
grlinecolors = ['#000000', '#555555', '#aaaaaa']
linecolors = ['#000000', '#3163e1', '#ff9b2f']

ax.plot(x, y, lw=1.5, color='k', label='func')
ax.plot(x_3, y_3, lw=1.5, color='r', label='3 order')
ax.plot(x_4, y_4, lw=1.5, color='g', label='4 order')
ax.plot(x_5, y_5, lw=1.5, color='b', label='5 order')
ax.plot(x_6, y_6, lw=1.5, color='m', label='6 order')


#ax.axis('equal')
plt.grid(True, linewidth = 0.5)
plt.tight_layout()
plt.legend()
plt.show()
```
